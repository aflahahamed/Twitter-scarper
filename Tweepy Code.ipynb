```python
#defined a fucntion to make a dataframe 
def extract_timeline_as_df (timeline_list):
    columns=set()
    allowed_type=[int,str]
    tweets_data=[]


    for status in timeline_list:
        #print(vars(status))
        #print(status.text)
        status_dict=dict(vars(status))
        keys = status_dict.keys()
        single_tweet_data={"User": status.user.screen_name,"author": status.author.screen_name}
        for k in keys:
            try:
                v_type = type(status_dict[k])
            except:
                v_type = None
            if v_type != None:
                if v_type in allowed_type:
                    single_tweet_data[k] = status_dict[k]
                    columns.add(k)
        tweets_data.append(single_tweet_data)
    header_cols = list(columns)
    header_cols.append("User")
    header_cols.append("author")
    df= pd.DataFrame(tweets_data,columns=header_cols)
    
    return df
```

# Get data from a hashtag

```python
#define a hashtag to search
# - helps us to filter the tweets containing those words in them
query = "#WonderWoman1984 -jQuery -從神話來的女神 -RT "
```

```python
tweets_dataa=[]
for i , status in enumerate(tweepy.Cursor(api.search, q= query).items(50)):
    single_tweet={"User": status.author.screen_name,'Text': status.text}
    tweets_dataa.append(single_tweet)
    #print(i,"author = " + status.author.screen_name,status.text )
#print(tweets_dataa)
```
